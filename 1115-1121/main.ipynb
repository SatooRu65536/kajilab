```python
from mcp_persor import BVHparser, plot
```

```python
bvhp = BVHparser('log/twister-b-1.bvh')
motion_df = bvhp.get_motion_df()
```

```python
joint_name = 'root'
motion_df = bvhp.get_joint_motion_df(joint_name, mode='absolute')

xlim = (0, 220)
ylim = (0, 0)

plot(
    df=motion_df,
    heads=[
        ['time', 'Xposition'],
        ['time', 'Yposition'],
        ['time', 'Zposition']
    ],
    title=f'{joint_name} の位置',
    xlabel='time [s]',
    ylabel='position [cm]',
    xlim=xlim,
    ylim=ylim,
    grid=True,
)

plot(
    df=motion_df,
    heads=[
        ['time', 'Xrotation'],
        ['time', 'Yrotation'],
        ['time', 'Zrotation']
    ],
    title=f'{joint_name} の回転',
    xlabel='time [s]',
    ylabel='rotation [deg]',
    xlim=xlim,
    ylim=ylim,
    grid=True,
)
```
